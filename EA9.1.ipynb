```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Trainingsdaten
qm = np.array([480,230,540,800,650,320,500,510,280,700,440,670,450,490]).reshape((-1,1))
preis = np.array([180,90,210,350,240,120,200,230,100,280,190,270,300,290])
```

```python
# Scatterplot der Trainingsdaten
plt.scatter(qm, preis)
plt.xlabel("qm")
plt.ylabel("Preis in Tausend")
plt.title("Grundstückspreise bezüglich Grundstücksgröße");
```

```python
# Linear Regression auf Trainingsdaten angewandt
regr = linear_model.LinearRegression()
regr.fit(qm, preis);
```

```python
# Testdaten zum Vorhersagen
qm_test = np.array([414, 379, 243, 462, 643, 226, 877, 626, 712, 694, 616, 337, 580, 890, 840, 394, 307, 681, 332, 670]).reshape((-1,1))
preis_test = np.array([170,150,80,190,230,90,360,280,330,300,270,150,240,400,370,195,130,265,150,250])
```

```python
# Vorhergesagter Preis
preis_pred = regr.predict(qm_test)
```

```python
# Lineare Regression der Testdaten
plt.scatter(qm_test, preis_test,  color='red')
plt.plot(qm_test, preis_pred, linewidth=2)
plt.xlabel("qm")
plt.ylabel("Preis in Tausend")
plt.title("Grundstückspreise bezüglich Grundstücksgröße");
```
